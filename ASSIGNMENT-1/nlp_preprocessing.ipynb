Task 2:Build Advanced Preprocessing Function
Implement the following function:
def preprocess_text(text):
    pass


```python
import re

# Helper: Normalize repeated characters (controlled)
def normalize_repeated_chars(word):
    return re.sub(r"(.)\1{2,}", r"\1\1", word)

# Main preprocessing function
def preprocess_text(text):

    # Error handling
    if not isinstance(text, str) or text.strip() == "":
        return {
            "tokens": [],
            "clean_sentence": ""
        }

    # Remove URLs & emails
    text = re.sub(r"(https?://\S+|www\.\S+|\S+@\S+)", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Lowercase
    text = text.lower()

    # Remove emojis & special chars
    text = re.sub(r"[^\w\s]", " ", text)

    # Normalize repeated characters
    words = text.split()
    words = [normalize_repeated_chars(w) for w in words]

    # Remove extra spaces
    text = " ".join(words)
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    # Custom filtering (keep meaningful small words)
    exceptions = {"no", "not"}
    filtered_tokens = []

    for token in tokens:
        if len(token) > 2 or token in exceptions:
            filtered_tokens.append(token)

    clean_sentence = " ".join(filtered_tokens)

    return {
        "tokens": filtered_tokens,
        "clean_sentence": clean_sentence
    }
```

Task 4: Token Analytics
For each processed sentence, compute:
Total number of tokens
Number of unique tokens
Average token length


```python
def analyze_tokens(tokens):
    if not tokens:
        return {
            "total": 0,
            "unique": 0,
            "avg_length": 0
        }

    return {
        "total": len(tokens),
        "unique": len(set(tokens)),
        "avg_length": sum(len(t) for t in tokens) / len(tokens)
    }


for sentence in test_data:
    result = preprocess_text(sentence)
    stats = analyze_tokens(result["tokens"])

    print(f"Sentence: {sentence}")
    print(f"Total Tokens: {stats['total']}")
    print(f"Unique Tokens: {stats['unique']}")
    print(f"Average Length: {stats['avg_length']:.2f}")
    print("-"*50)
```

Task 5: Frequency Analysis
Combine all tokens and use:
from collections import Counter


```python
from collections import Counter

all_tokens = []

for sentence in test_data:
    result = preprocess_text(sentence)
    all_tokens.extend(result["tokens"])

freq = Counter(all_tokens)

print("Top 10 Frequent Words:")
for word, count in freq.most_common(10):
    print(word, ":", count)

print("\nLeast 5 Frequent Words:")
least_common = sorted(freq.items(), key=lambda x: x[1])[:5]

for word, count in least_common:
    print(word, ":", count)
```

Task 6: Build Full Pipeline
def full_pipeline(text_list):
    pass


```python
def full_pipeline(text_list):
    final_tokens = []
    clean_sentences = []

    for text in text_list:
        result = preprocess_text(text)

        final_tokens.extend(result["tokens"])
        clean_sentences.append(result["clean_sentence"])

    return {
        "tokens": final_tokens,
        "clean_sentences": clean_sentences
    }
```
